```python
def get_type(hand: str):
    unique_chars = set(hand)

    if len(unique_chars) == 1:
        return 'Five of a kind'
    elif len(unique_chars) == 5:
        return 'High card'
    elif len(unique_chars) == 4:
        return 'One pair'
    elif len(unique_chars) == 2:
        max_repeats = 0
        for char in unique_chars:
            if hand.count(char) > max_repeats:
                max_repeats = hand.count(char)
        if max_repeats == 4:
            return 'Four of a kind'
        else:
            return 'Full house' 
    else:
        max_repeats = 0
        for char in unique_chars:
            if hand.count(char) > max_repeats:
                max_repeats = hand.count(char)
        if max_repeats == 3:
            return 'Three of a kind'
        else:
            return 'Two pair' 
    
get_type('TTT98')
```

```python
# Modify cards to sort by relative strength
def get_strength(hand:str):
    return hand.replace('T','B').replace('J','C').replace('Q','D').replace('K','E').replace('A','F')

get_strength('123456789TTTJJJQQQKKKAAA')
```

```python
types = {'Five of a kind':7, 'Four of a kind':6, 'Full house':5, 'Three of a kind':4, 'Two pair':3, 'One pair':2, 'High card':1}
```

```python
# Find the type of each hand
df[1] = df[0].apply(lambda x: get_type(x))

df
```

```python
#df = pd.read_csv('part 1 sample.csv',delimiter=' ', header=None)
df = pd.read_csv('part 1 input.csv',delimiter=' ', header=None)
df
```

```python
df.columns = ['hand', 'bid']
df
```

```python
df['type'] = df['hand'].apply(lambda x: get_type(x))
df
```

```python
df['type_rank'] = df['type'].apply(lambda x: types[x])
df
```

```python
df['sorting_hand'] = df['hand'].apply(lambda x: get_strength(x))
df
```

```python
df.sort_values(by=['type_rank','sorting_hand'], ascending=[True, True], inplace=True)
```

```python
df['rank'] = range(1, len(df) + 1)
df
```

```python
df['winnings'] = df['bid'] * df['rank']
df
```

```python
total = df['winnings'].sum()
total
```
